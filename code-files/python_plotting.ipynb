Import pandas.

```python
import pandas as pd
```

Read the data as a pandas DataFrame.

```python
accelerometer_df = pd.read_csv('data/accelerometer.csv', index_col=0) 
```

Check the data. (It is good practice.)

```python
accelerometer_df.head()
```

Plot the DataFrame

```python
plot = accelerometer_df.plot(title='Accelerometer Data')
```

```python
plot_x = accelerometer_df['x'].plot(title='Accelerometer Data (x)')
```

```python
plot_y = accelerometer_df['y'].plot(title='Accelerometer Data (y)')
```

```python
plot_z = accelerometer_df['z'].plot(title='Accelerometer Data (z)')
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=3)

accelerometer_df['x'].plot(ax=axes[0])
accelerometer_df['y'].plot(ax=axes[1], color='red')
accelerometer_df['z'].plot(ax=axes[2], color='green')

fig.suptitle('Accelerometer Data')
fig.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharey=True)
ax1.plot(accelerometer_df['x'])
ax1.set_title('x')

ax2.plot(accelerometer_df['y'], color='red')
ax2.set_title('y')

ax3.plot(accelerometer_df['z'], color='green')
ax3.set_title('z')

fig.suptitle('Accelerometer Data')

plt.show()
```

## Plotting the Light Sensor Data

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True)
ax1.plot(accelerometer_df['x'])
ax1.set_title('x')
ax1.set_xlabel('Time (ms)')

ax2.plot(accelerometer_df['y'], color='red')
ax2.set_title('y')
ax2.set_xlabel('Time (ms)')

ax3.plot(accelerometer_df['z'], color='green')
ax3.set_title('z')
ax3.set_xlabel('Time (ms)')

fig.suptitle('Accelerometer Data')

plt.show()
```
